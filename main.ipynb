```python
import numpy as np
import pickle as pkl
import json

prop = json.load(open("checkpoints/step/15-5-voc/topK-memory.json", 'r'))
rand = json.load(open("checkpoints/step/15-5-voc/COS-mixup-manifold-4_0.json", 'r'))
```

```python
prop_imgs = {}
rand_imgs = {}

idxes = np.load("data/voc/split/train_list.npy")

for cl, ids in prop.items():
    prop_imgs[cl] = [(f'data/voc/dataset/{idxes[id, 0]}', f'data/voc/dataset/{idxes[id, 1]}') for id in ids]
for cl, ids in rand.items():
    rand_imgs[cl] = [(f'data/voc/dataset/{idxes[id, 0]}', f'data/voc/dataset/{idxes[id, 1]}') for id in ids]
prop_imgs
```

```python
from PIL import Image
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.imshow(np.array(Image.open(prop_imgs['1'][0])))
plt.subplot(1, 2, 2)
plt.imshow(np.array(Image.open(rand_imgs['1'][0])))
plt.show()
```
